```
# Definindo os tipos de quarto e seus preços por noite
precos_quartos = {
    'simples': 100,  # Preço por noite para quarto simples
    'duplo': 150,    # Preço por noite para quarto duplo
    'luxo': 300      # Preço por noite para quarto luxo
}

# Lista para armazenar as reservas
reservas = []

# Função para adicionar uma reserva
def adicionar_reserva():
    # Solicita ao usuário o nome do hóspede
    nome = input("Digite o nome do hóspede: ")
    # Solicita ao usuário o tipo de quarto
    tipo_quarto = input("Digite o tipo de quarto (simples, duplo, luxo): ")
    # Solicita ao usuário a duração da estadia em noites
    duracao = int(input("Digite a duração da estadia (em noites): "))

    # Verifica se o tipo de quarto é válido
    if tipo_quarto in precos_quartos:
        # Calcula o custo total da estadia
        custo_total = precos_quartos[tipo_quarto] * duracao
        # Cria um dicionário com os detalhes da reserva
        reserva = {
            'nome': nome,
            'tipo_quarto': tipo_quarto,
            'duracao': duracao,
            'custo_total': custo_total
        }
        # Adiciona a reserva à lista de reservas
        reservas.append(reserva)
        print(f"Reserva adicionada: {reserva}")
    else:
        # Informa ao usuário que o tipo de quarto é inválido
        print("Tipo de quarto inválido.")

# Função para remover uma reserva pelo nome do hóspede
def remover_reserva():
    # Solicita ao usuário o nome do hóspede cuja reserva deve ser removida
    nome = input("Digite o nome do hóspede para remover a reserva: ")
    global reservas
    # Filtra a lista de reservas para remover a reserva do hóspede especificado
    reservas = [reserva for reserva in reservas if reserva['nome'] != nome]
    print(f"Reserva de {nome} removida.")

# Função para listar todas as reservas
def listar_reservas():
    # Itera sobre a lista de reservas e imprime cada uma
    for reserva in reservas:
        print(reserva)

# Loop principal do menu
while True:
    # Exibe o menu de opções
    print("\nMenu:")
    print("1. Adicionar reserva")
    print("2. Remover reserva")
    print("3. Listar reservas")
    print("4. Sair")

    # Solicita ao usuário que escolha uma opção
    escolha = input("Escolha uma opção: ")

    # Executa a função correspondente à escolha do usuário
    if escolha == '1':
        adicionar_reserva()
    elif escolha == '2':
        remover_reserva()
    elif escolha == '3':
        listar_reservas()
    elif escolha == '4':
        # Sai do loop e termina o programa
        break
    else:
        # Informa ao usuário que a opção é inválida
        print("Opção inválida. Tente novamente.")
```

```
# Lista para armazenar os carros no estacionamento
estacionamento = []

# Função para adicionar um carro
def adicionar_carro():
    # Solicita ao usuário a marca do carro
    marca = input("Digite a marca do carro: ")
    # Solicita ao usuário a cor do carro
    cor = input("Digite a cor do carro: ")
    # Solicita ao usuário a placa do carro
    placa = input("Digite a placa do carro: ")

    # Cria um dicionário com os detalhes do carro
    carro = {
        'marca': marca,
        'cor': cor,
        'placa': placa
    }
    # Adiciona o carro à lista de estacionamento
    estacionamento.append(carro)
    print(f"Carro adicionado: {carro}")

# Função para remover um carro pela placa
def remover_carro():
    # Solicita ao usuário a placa do carro a ser removido
    placa = input("Digite a placa do carro para remover: ")
    global estacionamento
    # Filtra a lista de estacionamento para remover o carro com a placa especificada
    estacionamento = [carro for carro in estacionamento if carro['placa'] != placa]
    print(f"Carro com placa {placa} removido.")

# Função para listar todos os carros no estacionamento
def listar_carros():
    # Itera sobre a lista de estacionamento e imprime cada carro
    for carro in estacionamento:
        print(carro)

# Loop principal do menu
while True:
    # Exibe o menu de opções
    print("\nMenu:")
    print("1. Adicionar carro")
    print("2. Remover carro")
    print("3. Listar carros")
    print("4. Sair")

    # Solicita ao usuário que escolha uma opção
    escolha = input("Escolha uma opção: ")

    # Executa a função correspondente à escolha do usuário
    if escolha == '1':
        adicionar_carro()
    elif escolha == '2':
        remover_carro()
    elif escolha == '3':
        listar_carros()
    elif escolha == '4':
        # Sai do loop e termina o programa
        break
    else:
        # Informa ao usuário que a opção é inválida
        print("Opção inválida. Tente novamente.")
```

```
# Dicionário para armazenar os projetos e suas tarefas
projetos = {}

# Função para adicionar um projeto
def adicionar_projeto():
    nome_projeto = input("Digite o nome do projeto: ")
    if nome_projeto not in projetos:
        projetos[nome_projeto] = []
        print(f"Projeto '{nome_projeto}' adicionado.")
    else:
        print(f"Projeto '{nome_projeto}' já existe.")

# Função para adicionar uma tarefa a um projeto
def adicionar_tarefa():
    nome_projeto = input("Digite o nome do projeto: ")
    if nome_projeto in projetos:
        titulo = input("Digite o título da tarefa: ")
        descricao = input("Digite a descrição da tarefa: ")
        responsavel = input("Digite o responsável pela tarefa: ")
        status = input("Digite o status da tarefa (não iniciado, em andamento, concluído): ")

        tarefa = {
            'titulo': titulo,
            'descricao': descricao,
            'responsavel': responsavel,
            'status': status
        }
        projetos[nome_projeto].append(tarefa)
        print(f"Tarefa '{titulo}' adicionada ao projeto '{nome_projeto}'.")
    else:
        print(f"Projeto '{nome_projeto}' não encontrado.")

# Função para atualizar o status de uma tarefa
def atualizar_tarefa():
    nome_projeto = input("Digite o nome do projeto: ")
    if nome_projeto in projetos:
        titulo = input("Digite o título da tarefa a ser atualizada: ")
        for tarefa in projetos[nome_projeto]:
            if tarefa['titulo'] == titulo:
                novo_status = input("Digite o novo status da tarefa (não iniciado, em andamento, concluído): ")
                tarefa['status'] = novo_status
                print(f"Status da tarefa '{titulo}' atualizado para '{novo_status}'.")
                break
        else:
            print(f"Tarefa '{titulo}' não encontrada no projeto '{nome_projeto}'.")
    else:
        print(f"Projeto '{nome_projeto}' não encontrado.")

# Função para listar todas as tarefas de um projeto
def listar_tarefas():
    nome_projeto = input("Digite o nome do projeto: ")
    if nome_projeto in projetos:
        for tarefa in projetos[nome_projeto]:
            print(tarefa)
    else:
        print(f"Projeto '{nome_projeto}' não encontrado.")

# Loop principal do menu
while True:
    print("\nMenu:")
    print("1. Adicionar projeto")
    print("2. Adicionar tarefa")
    print("3. Atualizar tarefa")
    print("4. Listar tarefas")
    print("5. Sair")

    escolha = input("Escolha uma opção: ")

    if escolha == '1':
        adicionar_projeto()
    elif escolha == '2':
        adicionar_tarefa()
    elif escolha == '3':
        atualizar_tarefa()
    elif escolha == '4':
        listar_tarefas()
    elif escolha == '5':
        break
    else:
        print("Opção inválida. Tente novamente.")
```
